```python
df = df[df['Period End'] != 'Weekly'].copy()

df['Period End'] = pd.to_datetime(df['Period End'], format='%d-%b-%Y')
```

```python
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

## Fuel Price Trend Over Time (National Average)

```python
df.plot(x='Period End', y='Pakistan')
figsize=(12, 6)
title=('National Average Petrol Price Over Time')
```

## City-wise Fuel Price Comparison (latest week)

```python
latest = df[df['Period End'] == df['Period End'].max()].drop(columns=['Period End'])
latest.T.plot(kind='bar', figsize=(14, 6), legend=False, title='City-wise Petrol Prices on ' + str(df['Period End'].max()))
```

## Average Petrol Price per City (2013–2025)

```python
avg_prices = df.drop(columns=['Period End']).mean().sort_values(ascending=False)
avg_prices.plot(kind='bar', figsize=(14, 6), title='Average Petrol Price by City (2013–2025)')
```

## Top 5 Cities with Highest Price Fluctuations

```python
price_fluctuations = df.drop(columns=['Period End']).std().sort_values(ascending=False)
price_fluctuations.head(5).plot(kind='bar', title='Top 5 Most Volatile Cities')
```
